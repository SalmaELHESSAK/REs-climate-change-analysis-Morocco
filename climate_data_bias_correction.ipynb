```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np

# Load the NetCDF files
historical_file = '/content/drive/MyDrive/data/MPI/temp/RF/tas_Morocco-22_MPI-M-MPI-ESM-MR_historical_r1i1p1_ICTP-RegCM4-7_v0_day_1970-2005.nc'
scenario1_file = '/content/drive/MyDrive/data/MPI/temp/RCP85/tas_Morocco-22_MPI-M-MPI-ESM-MR_rcp85_r1i1p1_ICTP-RegCM4-7_v0_day_2005-2100.nc'
scenario2_file ='/content/drive/MyDrive/data/MPI/temp/RCP26/tas_Morocco-22_MPI-M-MPI-ESM-MR_rcp26_r1i1p1_ICTP-RegCM4-7_v0_day_2005-2100.nc'
observed_file = '/content/drive/MyDrive/data/era5/era5_daily_2m_temperature_data_1979-2023.nc'

# Open datasets
historical_data = xr.open_dataset(historical_file)
scenario1_data = xr.open_dataset(scenario1_file)
scenario2_data = xr.open_dataset(scenario2_file)
observed_data = xr.open_dataset(observed_file)

# Align the time periods of historical data and observed data
time_slice = slice('1979-01-01', '2005-11-30')
historical_data = historical_data.sel(time=time_slice)
observed_data = observed_data.sel(time=time_slice)

# Print the time coordinates to check for alignment
print("Historical data time range:", historical_data['time'].values)
print("Observed data time range:", observed_data['time'].values)

# Ensure latitude and longitude ranges match
lat_range = slice(np.array(historical_data["latitude"]).max(),
                  np.array(historical_data["latitude"]).min())
lon_range = slice(np.array(historical_data["longitude"]).min(),
                  np.array(historical_data["longitude"]).max())

# Crop the observed dataset using the coordinate labels
observed_data = observed_data.sel(latitude=lat_range, longitude=lon_range)

# Rename variable in historical data to match observed data
historical_data = historical_data.rename({'tas': 't2m'})
scenario1_data = scenario1_data.rename({'tas': 't2m'})
scenario2_data = scenario2_data.rename({'tas': 't2m'})

# Ensure that variable names match and print data types
print("Data type of historical_data['t2m']:", historical_data['t2m'].dtype)
print("Data type of observed_data['t2m']:", observed_data['t2m'].dtype)

# Resample datasets to monthly frequency for historical and observed data
historical_data_monthly = historical_data.resample(time='1M').mean()
observed_data_monthly = observed_data.resample(time='1M').mean()

# Align the datasets by reindexing
observed_data_monthly = observed_data_monthly.reindex_like(historical_data_monthly, method='nearest')

# Select a specific grid point
lat_idx = 10  # Example index
lon_idx = 10  # Example index

# Calculate the mean bias (model data - observed data) for the specific grid point
mean_bias = np.mean(historical_data_monthly['t2m'][:, lat_idx, lon_idx] - observed_data_monthly['t2m'][:, lat_idx, lon_idx])

# Print mean bias before correction
print("Mean Bias before correction at grid point ({}, {}): {}".format(lat_idx, lon_idx, mean_bias))

# Apply bias correction to historical data for the specific grid point
historical_data_corrected = historical_data_monthly['t2m'][:, lat_idx, lon_idx] - mean_bias

# Calculate the mean bias after correction for the specific grid point
mean_bias_corrected = np.mean(historical_data_corrected - observed_data_monthly['t2m'][:, lat_idx, lon_idx])

# Print mean bias after correction
print("Mean Bias after correction at grid point ({}, {}): {}".format(lat_idx, lon_idx, mean_bias_corrected))

# Resample scenario data to annual frequency
scenario1_data_annual = scenario1_data.resample(time='1Y').mean()
scenario2_data_annual = scenario2_data.resample(time='1Y').mean()

# Apply bias correction to scenario data for the specific grid point
scenario1_data_corrected = scenario1_data_annual['t2m'][:, lat_idx, lon_idx] - mean_bias
scenario2_data_corrected = scenario2_data_annual['t2m'][:, lat_idx, lon_idx] - mean_bias

# Plot historical observed data, model data, and bias for a specific grid point
plt.figure(figsize=(12, 6))
plt.plot(historical_data_monthly['time'], historical_data_monthly['t2m'][:, lat_idx, lon_idx], label='Historical Model Data')
plt.plot(observed_data_monthly['time'], observed_data_monthly['t2m'][:, lat_idx, lon_idx], label='Historical Observed Data')

plt.xlabel('Time')
plt.ylabel('Temperature (K)')
plt.title('Biases in Historical Climate Model Data vs. Observed Data')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(historical_data_monthly['time'], historical_data_monthly['t2m'][:, lat_idx, lon_idx], label='Historical Model Data')
plt.plot(historical_data_monthly['time'], historical_data_corrected, label='Bias Corrected Historical Data')

plt.xlabel('Time')
plt.ylabel('Temperature (K)')
plt.title('Bias Correction in Historical Climate Model Data vs. Observed Data')
plt.legend()
plt.show()

# Bias corrected historical vs observed data
plt.figure(figsize=(12, 6))
plt.plot(observed_data_monthly['time'], observed_data_monthly['t2m'][:, lat_idx, lon_idx], label='Historical Observed Data')
plt.plot(historical_data_monthly['time'], historical_data_corrected, label='Bias Corrected Historical Data')

plt.xlabel('Time')
plt.ylabel('Temperature (K)')
plt.title('Bias Corrected Historical Climate Model Data vs. Observed Data')
plt.legend()
plt.show()

# Plot scenario data before and after bias correction for a specific grid point
plt.figure(figsize=(12, 6))
plt.plot(scenario1_data_annual['time'], scenario1_data_annual['t2m'][:, lat_idx, lon_idx], label='Scenario1 Model Data (RCP85)')
plt.plot(scenario1_data_corrected['time'], scenario1_data_corrected, label='Bias Corrected Scenario1 Data (RCP85)')

plt.xlabel('Time')
plt.ylabel('Temperature (K)')
plt.title('Bias Correction in RCP 8.5 Data')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(scenario2_data_annual['time'], scenario2_data_annual['t2m'][:, lat_idx, lon_idx], label='Scenario2 Model Data (RCP26)')
plt.plot(scenario2_data_corrected['time'], scenario2_data_corrected, label='Bias Corrected Scenario2 Data (RCP26)')
plt.xlabel('Time')
plt.ylabel('Temperature (K)')
plt.title('Bias Correction in RCP 2.6 Data')
plt.legend()
plt.show()
```

```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import scipy.special as special

# File paths
historical_file_wind10 = '/content/drive/MyDrive/data/MPI/Wind_Near_surface/RF/sfcWind_Morocco-22_MPI-M-MPI-ESM-MR_historical_r1i1p1_ICTP-RegCM4-7_v0_day_1970-2005.nc'
scenario1_file_wind10 = '/content/drive/MyDrive/data/MPI/Wind_Near_surface/RCP85/sfcWind_Morocco-22_MPI-M-MPI-ESM-MR_rcp85_r1i1p1_ICTP-RegCM4-7_v0_day_2005-2100.nc'
scenario2_file_wind10 = '/content/drive/MyDrive/data/MPI/Wind_Near_surface/RCP26/sfcWind_Morocco-22_MPI-M-MPI-ESM-MR_rcp26_r1i1p1_ICTP-RegCM4-7_v0_day_2005-2100.nc'
observed_file_u_wind10 = '/content/drive/MyDrive/data/era5/era5_daily_10m_u_component_of_wind_data_1979-2023.nc'
observed_file_v_wind10 = '/content/drive/MyDrive/data/era5/era5_daily_10m_v_component_of_wind_data_1979-2023.nc'

# Open datasets
historical_data_wind10 = xr.open_dataset(historical_file_wind10)
scenario1_data_wind10 = xr.open_dataset(scenario1_file_wind10)
scenario2_data_wind10 = xr.open_dataset(scenario2_file_wind10)
observed_data_u_wind10 = xr.open_dataset(observed_file_u_wind10)
observed_data_v_wind10 = xr.open_dataset(observed_file_v_wind10)

# Align the time periods of historical data and observed data
time_slice = slice('1979-01-01', '2005-11-30')
historical_data_wind10 = historical_data_wind10.sel(time=time_slice)
observed_data_u_wind10 = observed_data_u_wind10.sel(time=time_slice)
observed_data_v_wind10 = observed_data_v_wind10.sel(time=time_slice)

# Ensure latitude and longitude ranges match
lat_range = slice(np.array(historical_data_wind10["latitude"]).max(),
                  np.array(historical_data_wind10["latitude"]).min())
lon_range = slice(np.array(historical_data_wind10["longitude"]).min(),
                  np.array(historical_data_wind10["longitude"]).max())

observed_data_u_wind10 = observed_data_u_wind10.sel(latitude=lat_range, longitude=lon_range)
observed_data_v_wind10 = observed_data_v_wind10.sel(latitude=lat_range, longitude=lon_range)

# Compute wind speed from u and v components
observed_data_wind10 = np.sqrt(observed_data_u_wind10['u10']**2 + observed_data_v_wind10['v10']**2)
observed_data_wind10 = xr.DataArray(observed_data_wind10,
                             coords={'time': observed_data_u_wind10['time'],
                                     'latitude': observed_data_u_wind10['latitude'],
                                     'longitude': observed_data_u_wind10['longitude']},
                             dims=['time', 'latitude', 'longitude'],
                             name='sfcWind')

# Resample datasets to monthly frequency
historical_data_wind10_monthly = historical_data_wind10.resample(time='1M').mean()
observed_data_wind10_monthly = observed_data_wind10.resample(time='1M').mean()

# Align datasets by reindexing
observed_data_wind10_monthly = observed_data_wind10_monthly.reindex_like(historical_data_wind10_monthly, method='nearest')

# Function to calculate Weibull parameters using method of moments
def calculate_weibull_params(data):
    mean = data.mean(dim='time')
    std_dev = data.std(dim='time')
    k = (mean / std_dev) ** (-1.086)
    c = mean / special.gamma(1 + 1/k)
    return k, c

# Calculate Weibull parameters for observed and historical data
k_obs, c_obs = calculate_weibull_params(observed_data_wind10_monthly)
k_hist, c_hist = calculate_weibull_params(historical_data_wind10_monthly)

# Apply quantile mapping for bias correction
def weibull_cdf(x, k, c):
    return 1 - np.exp(-(x / c) ** k)

def weibull_ppf(p, k, c):
    return c * (-np.log(1 - p)) ** (1 / k)

def quantile_mapping(data, k_obs, c_obs, k_hist, c_hist):
    p = weibull_cdf(data, k_hist, c_hist)
    corrected_data = weibull_ppf(p, k_obs, c_obs)
    return corrected_data

# Apply quantile mapping to historical data
historical_data_wind10_corrected = quantile_mapping(historical_data_wind10_monthly['sfcWind'], k_obs, c_obs, k_hist, c_hist)

# Resample scenario data to annual frequency
scenario1_data_wind10_annual = scenario1_data_wind10.resample(time='1Y').mean()
scenario2_data_wind10_annual = scenario2_data_wind10.resample(time='1Y').mean()

# Apply quantile mapping to scenario data
scenario1_data_wind10_corrected = quantile_mapping(scenario1_data_wind10_annual['sfcWind'], k_obs, c_obs, k_hist, c_hist)
scenario2_data_wind10_corrected = quantile_mapping(scenario2_data_wind10_annual['sfcWind'], k_obs, c_obs, k_hist, c_hist)

# Plot historical observed data, model data, and bias for a specific grid point
lat_idx = 26 # Example index
lon_idx = 26  # Example index

plt.figure(figsize=(12, 6))
plt.plot(historical_data_wind10_monthly['time'], historical_data_wind10_monthly['sfcWind'][:, lat_idx, lon_idx], label='Historical Model Data')
plt.plot(observed_data_wind10_monthly['time'], observed_data_wind10_monthly[:, lat_idx, lon_idx], label='Historical Observed Data')
plt.xlabel('Time')
plt.ylabel('Wind Speed at 10m (m/s)')
plt.title('Biases in Historical Climate Model Data vs. Observed Data (Wind Speed at 10m)')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(historical_data_wind10_corrected['time'], historical_data_wind10_corrected[:, lat_idx, lon_idx], label='Bias Corrected Historical Data')
plt.xlabel('Time')
plt.ylabel('Wind Speed at 10m (m/s)')
plt.title('Bias Correction in Historical Climate Model Data vs. Observed Data (Wind Speed at 10m)')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(observed_data_wind10_monthly['time'], observed_data_wind10_monthly[:, lat_idx, lon_idx], label='Historical Observed Data')
plt.plot(historical_data_wind10_corrected['time'], historical_data_wind10_corrected[:, lat_idx, lon_idx], label='Bias Corrected Historical Data')
plt.xlabel('Time')
plt.ylabel('Wind Speed at 10m (m/s)')
plt.title('Bias Corrected Historical Climate Model Data vs. Observed Data (Wind Speed at 10m)')
plt.legend()
plt.show()

# Plot scenario data before and after bias correction for a specific grid point
plt.figure(figsize=(12, 6))
plt.plot(scenario1_data_wind10_annual['time'], scenario1_data_wind10_annual['sfcWind'][:, lat_idx, lon_idx], label='Scenario1 Model Data (RCP85)')
plt.plot(scenario1_data_wind10_corrected['time'], scenario1_data_wind10_corrected[:, lat_idx, lon_idx], label='Bias Corrected Scenario1 Data (RCP85)')
plt.xlabel('Time')
plt.ylabel('Wind Speed at 10m (m/s)')
plt.title('Bias Correction in RCP 8.5 Data (Wind Speed at 10m)')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(scenario2_data_wind10_annual['time'], scenario2_data_wind10_annual['sfcWind'][:, lat_idx, lon_idx], label='Scenario2 Model Data (RCP26)')
plt.plot(scenario2_data_wind10_corrected['time'], scenario2_data_wind10_corrected[:, lat_idx, lon_idx], label='Bias Corrected Scenario2 Data (RCP26)')
plt.xlabel('Time')
plt.ylabel('Wind Speed at 10m (m/s)')
plt.title('Bias Correction in RCP 2.6 Data (Wind Speed at 10m)')
plt.legend()
plt.show()
```
